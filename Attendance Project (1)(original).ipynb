```python
att_stat=separated[['Name','Student Id','Batch Timing','Timestamp','Time']]
grp_tm=att_stat.groupby('Batch Timing')
grp_tm.get_group('11:00:00')
```

```python
att_stat['Time'] = pd.to_datetime(att_stat['Time'], errors='coerce')
```

```python
ash=input("Enter the Date eg.'yyyy-mm-dd' : ")
lok=input("Enter the start time of Batch eg:'hh:mm:ss' : ")

st_time=grp_tm.get_group(lok)[grp_tm.get_group(lok)['Timestamp']==ash]
en_time=input("Enter the end time of Batch eg:'hh:mm:ss' : ")


start_time = pd.to_datetime(lok).time()
end_time = pd.to_datetime(en_time).time()  


valid_start_time = pd.to_datetime(lok) - pd.Timedelta(minutes=5) 
valid_end_time = pd.to_datetime(en_time) + pd.Timedelta(minutes=5) 

print("Valid Attendance's starting time for : ",lok," Batch ", valid_start_time)
print("Valid Attendance's ending time for : ",en_time, "  Batch ", valid_end_time)


o=[]
for i in st_time['Time']:
    if i>=valid_start_time and i <= valid_end_time:
        a=("Valid Attendance : ",i)
        o.append(a)
    else:
        b=("Invalid Attendance : ",i)
        o.append(b)
        

bush=pd.DataFrame(o,columns=['Attendance Status','Attendance Time'])
hup=st_time[['Name','Student Id','Timestamp']]
bush.reset_index(drop=True, inplace=True)
hup.reset_index(drop=True, inplace=True)
hurr=pd.concat([hup,bush],axis=1)
hurr
```
