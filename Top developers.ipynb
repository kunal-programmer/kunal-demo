# live web scapping project

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
page_data=[]
extra_data=[]
for j in range(1,2):
    link="https://www.topdevelopers.co/directory/big-data-analytics-companies? page={}".format(j)
    page=requests.get(link)
    soup=BeautifulSoup(page.content,"html.parser")
    print(soup.prettify())
    y=soup.find_all(class_="col-xl-12 col-lg-12 comp_listing list_holder")
    d=soup.find_all("div",class_="media" )
    for i in range(len(d)):
        sample_c=d[i]
    if i==len(sample_c.find_all(class_="media-left img-thumbnail")):
        try:
            company_name=sample_c.select("h3 a")[0].text
            try:
                tag_line=sample_c.find(class_="d-xl-inline-flex d-md-block d-lg-block d-block align-self-center").find(class_="tag_line").text
            except:
                Tag_line= page_data.append([company_name,Tag_line])
        except:
            pass
```

```python
ddd=[i.text.strip("\n") for i in  y[0].find_all(class_="set_es_box")]
```

```python
fl=[]
for i in list(range(0,len(ddd),4)):
    md=ddd[i:i+4]
    fl.append(list(map(lambda x :x.strip(),md)))
for extra_details in fl:
    no_of_emp=extra_details[0]
    salary_hr=extra_details[1]
    project_budget=extra_details[2]
    country=extra_details[3]
    extra_data.append([no_of_emp,salary_hr,project_budget,country])
```

```python
finaldata=[page_data[i]+extra_data[i] for i in range(len(page_data))]
```

```python
df=pd.DataFrame(finaldata,columns="company_Name Tag_Line Total_Employee Salary Project_Budget Country".split())
```
